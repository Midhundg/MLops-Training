```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/industrial_sensor_dataset.csv')
df
```

**Task1**

```python
X = df[['RPM']]
y=df[['Temperature']]
```

```python
model = linear_model.LinearRegression()
model.fit(X,y)
print("Predicted Value:",model.predict([[3000]]))
y_pred = model.predict(X)
mse=mean_squared_error(y,y_pred)
print("Mean Squared Error:",mse)
```

**Task 2**

```python
X1 = df[['RPM','Vibration','Pressure']]
y1 = df[['Temperature']]
```

```python
model1 = linear_model.LinearRegression()
model1.fit(X1,y1)
print("Predicted Value:",model1.predict([[3000,3.75,51.16]]))
y1_pred = model1.predict(X1)
mse1=mean_squared_error(y1,y1_pred)
print("Mean Squared Eroor:",mse1)
```

**Task 3**

```python
import pickle

with open('model_pickle','wb') as file:
  pickle.dump(model,file)

with open('model_pickle','rb') as file:
  mp = pickle.load(file)

print("Predicted Value using Pickle for Linear Regression:",mp.predict([[3000]]))

import joblib
joblib.dump(model1,'model_joblib')
mj = joblib.load('model_joblib')
print("Predicted Value using Joblib for MultiLinear Regression:",mj.predict([[3000,3.75,51.16]]))
```
