```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'],dtype = int)
dummies
```

```python
df=pd.concat([df,dummies],axis = 'columns')
df
```

```python
merged = df.drop(['Car Model','Mercedez Benz C class'],axis = 'columns')
merged
```

```python
X = merged.drop(['Sell Price($)'],axis = 'columns')
y=merged['Sell Price($)']
model = linear_model.LinearRegression()
model.fit(X,y)
model.predict([[40000,10,0,0]])
```
