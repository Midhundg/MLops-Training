```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/homeprices-2.csv')
df
```

```python
x=df[['area']]
y=df['price']
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
model.predict([[3100]])
```

```python
import joblib
```

```python
joblib.dump(model,'model_joblib')
```

```python
mj=joblib.load('model_joblib')
```

```python
mj.predict([[3100]])
```
