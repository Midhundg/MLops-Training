```python
import pandas as pd
from sklearn import linear_model
df=pd.read_csv('/content/homeprices-2.csv')
df
```

```python
x=df[['area']]
y=df['price']
model=linear_model.LinearRegression()
model.fit(x,y)
model.predict([[3100]])
```

```python
import pickle

with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

```python
with open('model_pickle','rb') as file:
    mp=pickle.load(file)
```

```python
mp.predict([[3100]])
```
